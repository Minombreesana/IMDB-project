```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
movies = pd.read_csv('../Data/peliculas.csv')
movies.head()
```

```python
num = (movies.dtypes == float) | (movies.dtypes == 'int64')
num_cols = [c for c in num.index if num[c]]
num_cols
```

```python
movies_num = movies[num_cols]
movies_num.head()
```

```python
financials = pd.read_csv('../Data/thenumbers.csv')
financials.head()
```

```python
financials = financials[['movie_title', 'worldwide_gross', 'production_budget']]
financials.head()
```

```python
movies_num = pd.concat([movies_num, movies['movie_title']], axis = 1)
```

```python
movies_num.head()
```

```python
financials.head()
```

```python
movies_v2 = pd.merge(financials, movies_num, on = 'movie_title', how = 'left')
```

```python
movies_v2.head()
```

```python
movies_v2.duplicated(['movie_title'], keep = False).value_counts()
```

```python
movies_v3 = movies_v2.drop_duplicates(subset='movie_title')
```

```python
opening = pd.read_csv('../Data/opening.csv')
opening.head()
```

```python
movies_v4 = pd.merge(opening, movies_v3, on = 'movie_title', how = 'left' )
```

```python
movies_v4.duplicated(['movie_title'], keep = False).value_counts()
```

```python
movies_v4.notnull().apply(pd.Series.value_counts)
```

```python
available = ((movies_v4 != 0) & (movies_v4.notnull()))
available.all(axis = 1).value_counts()
```

```python
mask = available['worldwide_gross']
movies_v4 = movies_v4[mask]
movies_v4.head()
```

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(missing_values=np.nan, strategy = 'mean')
```

```python
movies_v4 = movies_v4.drop('movie_title', axis = 1)
```

```python
movies_v4 = movies_v4.drop('duration.1', axis = 1)
```

```python
values = imputer.fit_transform(movies_v4)
values
```

```python
x = pd.DataFrame(values)
```

```python
x.columns = movies_v4.columns
```

```python
x.index = movies_v4.index
```

```python
x = x.drop('gross', axis = 1)
```

```python
x.head()
```

```python
x = x.drop('Unnamed: 0', axis = 1)
```

```python
x.head()
```

```python
x.to_csv('../Data/Intermediate_results/X_opening.csv', index = False)
```
